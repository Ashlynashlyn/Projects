# Scraping Data from a Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://github.com/SimplifyJobs/Summer2024-Internships?tab=readme-ov-file#we-love-our-contributors-%EF%B8%8F%EF%B8%8F" 

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")
```

```python
table = soup.find_all("table")[1]
```

```python
table_titles = table.find_all('th')

table_titles = [title.text.strip() for title in table_titles]

print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)

df
```

```python
data = table.find_all("tr")
print(data)
```

```python
for row in data[1:]:
    row_data = row.find_all("td")
    row_data = [item.text.strip() for item in row_data]
    print(row_data)
    
    location = len(df)
    df.loc[location] = row_data
    
```
